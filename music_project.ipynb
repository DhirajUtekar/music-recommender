```python
new_data = data[[
    'id', 'name', 'artists', 'genres', 'text',
    'year', 'release_date', 'popularity',
    'duration_sec',
    'acousticness','danceability','energy','loudness',
    'instrumentalness','liveness','speechiness','tempo','valence',
    'mood_happy','mood_chill'
]].copy()
```

```python
new_data.head()
```

```python
new_data['text'] = (new_data['text'].astype(str).str.lower()
    .str.replace(r'[^a-z0-9 ]+', ' ', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

new_data['text'] = new_data['text'].astype(str).fillna('').str.strip()

tfidf = TfidfVectorizer(stop_words='english',max_features=5000)
```

```python
tfidf_matrix = tfidf.fit_transform(new_data['text'])
```

```python
tfidf_matrix.shape
```

```python
svd = TruncatedSVD(n_components=50, random_state=42)
tfidf_svd = svd.fit_transform(tfidf_matrix)
```

```python
tfidf_svd.shape
```

```python
tfidf.fit(new_data['text'])

feature_names = tfidf.get_feature_names_out()
components = svd.components_

for i, comp in enumerate(components):
    top_indices = comp.argsort()[-10:]
    top_words = feature_names[top_indices]
    print(f"\nSVD Component {i+1}:")
    print(top_words)
```

```python
from sklearn.neighbors import NearestNeighbors

nn = NearestNeighbors(metric='cosine', n_neighbors=11, algorithm='brute')
nn.fit(tfidf_svd)
```

```python
def recommend(i, top_n=10):
    distances, indices = nn.kneighbors(tfidf_svd[i].reshape(1,-1), n_neighbors=top_n+1)
    distances = distances.flatten()
    indices = indices.flatten()

    mask = indices != i
    indices = indices[mask][:top_n]
    distances = distances[mask][:top_n]

    return new_data.iloc[indices][['id','name','artists','genres']]
```

```python
recommend(50, top_n=5)
```
